```python
def petla(n):
  def koniec():
    return None

  def kontynuuj():
    print(n)
    return petla(n-1)

  return [koniec, kontynuuj][n>0]()
```

```python
petla(12)
```

```python
def petla_f(n):
  return (lambda: None,
          lambda: (print(n),petla(n-1)))[n>0]()
```

```python
petla_f(9)
```

```python
petla_pyth = lambda n: (lambda f: f(f,n))(
    lambda self, x: (lambda: None,
                     lambda: (print(x),self(self,x-1)))[x>0]())
```

```python
petla_pyth(9)
```
