```python
class Samochod:
  def __init__(self, marka, model, rok_produkcji, przebieg):
    self.marka = marka
    self.model = model
    self.rok_produkcji = rok_produkcji
    self.przebieg = przebieg

  def __new__(cls,*args,**kwargs):
    return super().__new__(Samochod)

  def __repr__(self):
    return f"Nowy obiekt klasy {self.__class__.__name__} o adresie {id(self)} -> {self.marka}"

  def jedz(self, km):
    self.przebieg += km

  def info(self):
    print(f"Samochód marki: {self.marka}, przebieg: {self.przebieg}")
```

```python
moj_sam = Samochod("BMW","czarny",2015,123900)
moj_sam.jedz(233)
moj_sam.info()
print(moj_sam)
```
