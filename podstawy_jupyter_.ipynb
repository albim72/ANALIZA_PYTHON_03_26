# Kolekcje języka **Python**: <span style="color:red; font-size:28pt;">lista</span>, <span style="color:blue; font-size:28pt;">krotka</span>, <span style="color:orange; font-size:28pt;">zbiór</span>, <span style="color:silver; font-size:28pt;">słownik</span>.

```python
#lista
liczby = [5,5,6,7,8,9,0,53,57,90,3,7.5,23.5,True]
print(liczby)
print(type(liczby))
liczby.append(88)
print(liczby)
```

```python
liczby.insert(5,101)
print(liczby)
print(liczby[7])
```

```python
#krotka
imiona = ("Jan","Henryk","Anna","Leon","Klaudia","Nadia","Jan")
print(imiona)
print(type(imiona))
```

```python
print(imiona.index("Henryk"))
print(imiona.count("Jan"))
```

```python
#zbiór
drzewa = {"sosna","dąb","osika","jodła","buk","wiśnia","dąb"}
print(drzewa)
print(type(drzewa))
```

```python
drzewa.add("jabłoń")
print(drzewa)
```

```python
drzewa.update(["świerk","jesion","cyprys"])
print(drzewa)
```

```python
print("sosna" in drzewa)
```

```python
nb = [5,3,7,8,0,3,2,4,5,3,2,2,46,7,32,45,3,3,5,24,6,7,2,2]
nb = list(set(nb))
print(nb)
```

```python
#słownik - (map) dictionary
osoba = {
    "imię":"Eryk",
    "nazwisko":"Kot",
    "miasto":"Kraków",
    "wiek":33,
    "wniosek":True
}
```

```python
print(osoba)
print(type(osoba))
```

```python
osoba["kolor"] = "czarny"
print(osoba)
osoba["miasto"] = "Lublin"
print(osoba)
```
